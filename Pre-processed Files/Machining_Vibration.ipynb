```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("maximilianfellhuber/cnc-machining-data")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

base_dir = r"C:\Users\krishsh\.cache\kagglehub\datasets\maximilianfellhuber\cnc-machining-data\versions\2"

data_records = []

for machine in ['M01', 'M02', 'M03']:
    machine_path = os.path.join(base_dir, machine)
    
    for i in range(15):
        op_folder = f'OP{i:02d}'
        op_path = os.path.join(machine_path, op_folder)

        if not os.path.exists(op_path):
            continue

        for condition in ['good', 'bad']:
            condition_path = os.path.join(op_path, condition)

            if not os.path.exists(condition_path):
                continue

            for file in os.listdir(condition_path):
                if file.lower().endswith(('.xlsx', '.xls', '.csv')):
                    file_path = os.path.join(condition_path, file)

                    try:
                        # Read file based on extension
                        if file.lower().endswith('.csv'):
                            df = pd.read_csv(file_path)
                        else:
                            df = pd.read_excel(file_path)

                        # Check if required columns exist
                        if all(col in df.columns for col in ['x', 'y', 'z']):
                            medians = df[['x', 'y', 'z']].median()

                            data_records.append({
                                'median_x': medians['x'],
                                'median_y': medians['y'],
                                'median_z': medians['z'],
                                'Operation_Number': op_folder,
                                'Machine_Number': machine,
                                'condition': condition,
                                'filename': file
                            })
                        else:
                            print(f"⚠️ Skipped (columns missing): {file_path}")
                    except Exception as e:
                        print(f"❌ Error reading {file_path}: {e}")

# Final DataFrame
final_df = pd.DataFrame(data_records)

# Save result
output_path = os.path.join(base_dir, 'cnc_filewise_medians.csv')
final_df.to_csv(output_path, index=False)

print(f"✅ Done! Data saved to: {output_path}")
```

```python
df=df.drop(columns='filename')
df
```

```python
df['condition'] = df['condition'].map({'good':1,'bad':0})
df
```

```python
operation_encoding = df.groupby('Operation_Number')['condition'].mean()
df['Operation_Number_encoded'] = df['Operation_Number'].map(operation_encoding)

machine_encoding = df.groupby('Machine_Number')['condition'].mean()
df['Machine_Number_encoded'] = df['Machine_Number'].map(machine_encoding)
```

```python
df.drop(columns=['Operation_Number', 'Machine_Number'], inplace=True)
```
