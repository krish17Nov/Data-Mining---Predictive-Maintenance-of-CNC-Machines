```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler

dataframes = {}

for i in range(1,19):
    file_name = f'experiment_{i:02}.csv'
    df_name = f'df_{i:02}'
    dataframes[df_name] = pd.read_csv(file_name)

```

```python
dataframes['df_01']
```

```python
features_to_exclude = 'Machining_Process'

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

features_to_scale = [col for col in dataframes['df_01'].columns if col!=features_to_exclude]
ss.fit(dataframes['df_01'][features_to_scale])

scaled_data = ss.transform(dataframes['df_01'][features_to_scale])

scaled_df = pd.DataFrame(scaled_data, columns = features_to_scale)
scaled_df[features_to_exclude] = dataframes['df_01'][features_to_exclude]
scaled_df
```
